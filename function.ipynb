```python
## QUESTION 2 >>  Explain the concept of function arguments and parameters in Python

#In Python, functions are blocks of reusable code designed to perform a specific task. When defining a function, you can specify parameters that act as placeholders for the values that the function will work with. When you call the function, you provide arguments that are passed to these parameters.

#Parameters

#Parameters are the variables listed inside the parentheses in a function definition. They are used to receive input values when the function is called. Think of them as placeholders that tell the function what kind of input it expects.

#Example of defining parameters:

def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

# In this example, name and age are parameters of the function greet.
```

```python
#Arguments
#Arguments are the actual values you provide to a function when calling it. They correspond to the parameters defined in the function. When you call the function, you supply values (arguments) that get assigned to the corresponding parameters.

#Example of calling the function with arguments:

greet("Alice", 30)

#In this case, "Alice" is the argument for the name parameter, and 30 is the argument for the age parameter.
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#Types of Arguments

#1 Positional Arguments: The most common type, where arguments are passed in the order of parameters.
def add(a, b):
    return a + b

print(add(5, 3))  # Output: 8
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#2 Keyword Arguments: You can specify the arguments by name when calling the function, allowing you to pass them in any order.
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet(age=25, name="Bob")  # Output: Hello, Bob! You are 25 years old.
#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
#3 Default Arguments: You can assign default values to parameters, making them optional when calling the function.
def greet(name, age=20):
    print(f"Hello, {name}! You are {age} years old.")

greet("Charlie")   # Output: Hello, Charlie! You are 20 years old.
greet("Dana", 25)  # Output: Hello, Dana! You are 25 years old.

#Variable-Length Arguments: Python allows functions to accept an arbitrary number of arguments using *args (for positional arguments) and **kwargs (for keyword arguments).

#*args: Collects extra positional arguments as a tuple.

def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3, 4))  # Output: 10

#**kwargs: Collects extra keyword arguments as a dictionary.

def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Eve", age=30)  # Output: name: Eve, age: 30


```

```python
#Summary

#Parameters are the variables defined in the function signature.
#Arguments are the values you pass when calling the function.
#You can use positional arguments, keyword arguments, default arguments, or variable-length arguments depending on your needs.
#This flexibility allows Python functions to be versatile and easily adaptable to different use cases.



```

```python
## questions 4 >> What is the purpose of the `return` statement in a Python function

#The return statement in a Python function serves the purpose of sending a result back to the caller. It allows the function to produce a value that can be used later in the program. Once the return statement is executed, the function stops executing, and the specified value is returned to the place where the function was called.

#Purpose of return:
#Exits the function: When the return statement is reached, the function stops executing immediately.
#Sends a value back: The value provided in the return statement is sent back to the caller, which can be assigned to a variable or used directly.
#Example: Using return to Return a Value from a Function

def multiply(a, b):
    result = a * b  # Multiply the two inputs
    return result   # Return the result to the caller

# Calling the function and storing the result

product = multiply(4, 5)

# Using the returned value
print(product)  # Output: 20

#Breakdown:
#The function multiply takes two parameters a and b.
#It calculates their product and stores the result in the variable result.
#The return result statement sends the value of result (which is 20 in this case) back to the caller.
#We store the returned value in the variable product and then print it.


#Without the return statement, the function would just perform the calculation but wouldn't give us access to the result.
```

```python
## question5>>  What are iterators in Python and how do they differ from iterables?

#In Python, iterators and iterables are related concepts but serve different purposes. Understanding the difference between them is important when working with loops and other operations that involve sequences of data.

#Iterable
#An iterable is any object in Python that can be iterated (looped) over. In other words, it is an object that provides an iterator when passed to the iter() function. Most common iterables are sequences like lists, tuples, strings, and dictionaries.
#An iterable is any object that implements the __iter__() method, which returns an iterator, or the __getitem__() method (like a list, which can be indexed).

#Iterator
#An iterator is an object that represents a stream of data and can be traversed (iterated) through. It is obtained from an iterable using the iter() function. An iterator must implement two methods:
#__iter__(): This method returns the iterator object itself.
#__next__(): This method returns the next item from the iterable. When there are no more items left, it raises the StopIteration exception.

#Key Differences
#Iterable: An object that can be iterated over (e.g., list, tuple, string). It can return an iterator.
#Iterator: An object that represents a sequence and knows how to traverse through it. It is the object that does the actual iteration.
#Example: Difference Between Iterable and Iterator

# Example with a list (which is an iterable)

# Create a list (iterable)
my_list = [1, 2, 3, 4]

# Get an iterator from the list
my_iterator = iter(my_list)

# Now, we can manually iterate using the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4

# If we try to go beyond the elements, it will raise StopIteration
try:
    print(next(my_iterator))  # Raises StopIteration
except StopIteration:
    print("No more items!")

#Explanation:
#Iterable: The my_list object is an iterable, meaning it can be passed to the iter() function to produce an iterator.
#Iterator: The my_iterator is an iterator, which is created by calling iter(my_list). It knows how to produce the next item in the sequence using the next() function. Once it has exhausted all the items, calling next() again raises the StopIteration exception.

#Summary
#Iterable: An object that can be passed to iter() to create an iterator.
#Iterator: An object that maintains the state of iteration and produces the next item when next() is called.
#In simple terms, iterables are objects that can provide iterators (like lists, tuples, strings), and iterators are objects that track the progress of iteration over an iterable.
```

```python
## question 6>>  Explain the concept of generators in Python and how they are defined

#Generators in Python
#A generator in Python is a special type of iterable that allows you to iterate over a sequence of values lazily (i.e., one at a time) rather than storing all the values in memory at once. This makes generators more memory-efficient than lists, especially when dealing with large data sets or streams of data.

#Key Concepts of Generators

#Lazy Evaluation:
#Generators don’t generate all their items at once. Instead, they yield items one at a time when requested. This means that values are produced only when needed, reducing memory usage.

#State Retention:
#Generators maintain their state between successive calls. When the generator yields a value, it "pauses" its execution and can resume later from where it left off.

#Memory Efficiency:
#Generators are memory-efficient because they don’t store the entire sequence in memory. They only compute one value at a time as you iterate over them.


#How to Define a Generator
#Generators are defined using two primary methods:

#Generator Functions (using the yield keyword).
#Generator Expressions (using a compact syntax similar to list comprehensions).

#1. Generator Function
#A generator function is a function that contains one or more yield statements. Each time the function is called, it produces a value and pauses, maintaining its state. When you call next() on the generator, it resumes from where it left off and continues to yield values until the function is exhausted.

#Example of a Generator Function:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yield the current count and pause
        count += 1

# Create a generator object
counter = count_up_to(5)

# Use the generator
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
print(next(counter))  # Output: 4
print(next(counter))  # Output: 5

# After the generator is exhausted, it raises StopIteration
# print(next(counter))  # Raises StopIteration exception



#Explanation of the Example:
#The count_up_to function is a generator function. It starts a while loop that yields the value of count each time it's called.
#The yield statement pauses the function, returning the value of count to the caller.
#Every time next(counter) is called, the generator resumes from where it left off, incrementing count and yielding the next value.
#After all values have been yielded, calling next() again raises a StopIteration exception, signaling that the generator is exhausted.


#2. Generator Expression
#A generator expression is a more concise way to create generators. It is similar to a list comprehension but uses parentheses () instead of square brackets [].

#Example of a Generator Expression:

# A generator expression to generate squares of numbers
squares = (x * x for x in range(1, 6))

# Use the generator
print(next(squares))  # Output: 1
print(next(squares))  # Output: 4
print(next(squares))  # Output: 9
print(next(squares))  # Output: 16
print(next(squares))  # Output: 25


#Explanation of the Example:
#The expression (x * x for x in range(1, 6)) is a generator expression that yields the square of each number from 1 to 5.
#Each call to next(squares) retrieves the next value in the sequence (the square of x).
#After the generator has yielded all values, calling next(squares) again would raise a StopIteration exception.

#Advantages of Generators:
#Memory Efficiency: Since generators only generate values one at a time, they don't store the entire sequence in memory. This is especially useful when working with large datasets or streams.
#Lazy Evaluation: You can work with potentially infinite sequences without loading the entire sequence into memory.
#Improved Performance: For large datasets, generators can reduce memory usage and increase performance by yielding values only when needed.

#Summary
#Generators are a type of iterable that produce values one at a time using the yield statement.
#They are defined either using a generator function (with yield) or a generator expression (with parentheses).
#Generators are memory-efficient and support lazy evaluation, making them ideal for handling large datasets or infinite sequences.

#In short, generators allow you to iterate over a sequence of data without the overhead of storing the entire sequence in memory, yielding one item at a time when requested.
```

```python
## question 8 >> What is a lambda function in Python and when is it typically used

#A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike a regular function defined with def, a lambda function is a one-liner that can accept any number of arguments but can only have one expression.

#Syntax:
# lambda arguments: expression

#arguments: This is the input to the lambda function (it can be zero or more arguments).
#expression: This is the single expression that the lambda evaluates and returns.
#Lambda functions are often used in situations where you need a simple function for a short period, especially when you don't want to define a full function with a def block.

#When is it typically used?
#Lambda functions are commonly used in the following scenarios:

#Passing functions as arguments: Lambda functions are often used when you need to pass a simple function to functions like map(), filter(), sorted(), etc.
#Short-lived functions: When you need a quick function for a specific purpose and don’t want to create a separate, named function.
#Functional programming: They are useful in functional programming paradigms where you often pass functions around as first-class citizens.

#Example:
#Let’s use a lambda function with sorted(), which sorts a list of tuples based on the second element of each tuple.

#Problem:
#You have a list of tuples, and you want to sort them by the second element of each tuple.

# List of tuples
data = [(1, 'apple'), (3, 'orange'), (2, 'banana')]

# Sorting using a lambda function to sort by the second element (index 1)
sorted_data = sorted(data, key=lambda x: x[1])

# Output the sorted list
print(sorted_data)

#Explanation:
#The lambda function lambda x: x[1] takes an element x (which is a tuple) and returns the second element (x[1]), which is used as the sorting key.
#The sorted() function uses this lambda to sort the list of tuples based on the second element (the fruit name in this case).

#Why use a lambda here?
#Concise: The lambda function is a compact way to define the sorting criterion in a single line, without having to define a separate named function using def.
#Convenient for one-time use: The lambda is used just for the specific task of sorting in this context, so there’s no need to clutter the code with a function definition.

#In Summary:
#A lambda function is a short, anonymous function defined with lambda.
#It’s commonly used for quick, throwaway functions, especially when passed as arguments to functions like sorted(), map(), or filter().
#Lambdas are best used when the function is simple and doesn’t need to be reused elsewhere
```

```python
## question 9>>  Explain the purpose and usage of the `map()` function in Python.

#The map() function in Python is used to apply a given function to all the items in an iterable (like a list or tuple) and returns a map object (which is an iterator) containing the results. Essentially, it allows you to transform data in an iterable in a concise and efficient way.

#Purpose:
#Transformation: map() is used to transform all elements in an iterable based on a function you provide.
#Efficiency: It is often more efficient and concise than using a loop to apply a function to each item in a collection.
#Functional programming: map() is a functional programming tool that applies a function to each item in an iterable, returning the results as a map object (which can be converted to a list or other iterable).

#Syntax:
#map(function, iterable)

#function: The function that is applied to each item in the iterable. This can be a regular function or a lambda function.
#iterable: An iterable (e.g., list, tuple) whose elements will be transformed by the function.

#Returns:
#A map object (an iterator), which can be converted into other types like a list, tuple, etc.

#Example:
#Let's say we have a list of numbers, and we want to square each number in the list. We can use map() to apply a function that squares each element.

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Function to square a number
def square(x):
    return x * x

# Using map to apply the square function to each element in the list
squared_numbers = map(square, numbers)

# Convert the map object to a list to see the result
squared_numbers_list = list(squared_numbers)

# Output the squared numbers
print(squared_numbers_list)


#Explanation:
#The square function is applied to each element in the numbers list.
#The map() function returns a map object, which is an iterator.
#We convert the map object into a list to see the results of applying the square function to each element.

#Using a Lambda with map():
#You can also use a lambda function with map() for more concise code. For example, instead of defining a separate square() function, you can directly pass a lambda to map().

# Using lambda with map to square the numbers
squared_numbers = map(lambda x: x * x, numbers)

# Convert to a list and print
print(list(squared_numbers))

#Explanation:
#The lambda function lambda x: x * x is directly passed to map(), which applies it to each element of the numbers list.

#Key Points:
#Convenience: map() is a more concise and functional way to transform all items in an iterable, as opposed to writing a loop.
#Efficiency: map() returns an iterator, so it’s memory efficient when working with large datasets (it doesn’t generate all values at once).
#Use Case: Ideal for cases where you want to apply the same operation to each element in an iterable, like modifying or transforming the data.

```

```python
#question 10>>  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

#In Python, map(), reduce(), and filter() are all higher-order functions that apply a function to elements in an iterable, but they differ in their purpose, behavior, and output. Here's a breakdown of the differences, followed by an example for each function.

#1. map()
#Purpose: map() is used to apply a function to each element in an iterable and return a new iterable (like a list or iterator) with the results.
#Output: Returns an iterator (or map object in Python 3), which you can convert to other collections like a list or tuple.
#Use Case: When you want to transform or modify each element in a collection.

#Syntax:
######## map(function, iterable)


#2. reduce()
#Purpose: reduce() applies a function cumulatively to the items of an iterable (from left to right) to reduce them to a single value.
#Output: Returns a single value.
#Use Case: When you need to aggregate elements, e.g., summing or multiplying all elements in an iterable.
#Syntax:
from functools import reduce
#reduce(function, iterable)


#3. filter()
#Purpose: filter() is used to filter out elements from an iterable based on a condition. It applies a function that returns either True or False for each element, keeping only the elements for which the function returns True.
#Output: Returns an iterator of the elements that evaluate to True in the function.
#Use Case: When you want to filter elements from a collection based on a specific condition.
#Syntax:
#filter(function, iterable)


#>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

#Examples:


#Let’s use a list of numbers [1, 2, 3, 4, 5] and demonstrate how each of these functions works.

#1. Using map() (Transform each element)
#Suppose we want to square each number in the list:

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using map to square each number
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert to a list and print
print(list(squared_numbers))

## Explanation: map() applies the lambda function (which squares each number) to every element in the list, returning a new iterable with the squared values.

#2. Using reduce() (Aggregate elements)
#Now, let’s use reduce() to calculate the sum of the numbers in the list:

from functools import reduce

# Using reduce to sum all numbers
sum_of_numbers = reduce(lambda x, y: x + y, numbers)

# Print the result
print(sum_of_numbers)

#Explanation: reduce() cumulatively applies the lambda function (lambda x, y: x + y) to the list, summing the elements. It starts with the first two elements (1 and 2), then applies the function to the result and the next element, and so on, until it processes all elements, resulting in 15.

#3. Using filter() (Filter elements based on condition)
#Let’s use filter() to filter out even numbers from the list:
#Using filter to get only even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)

# Convert to a list and print
print(list(even_numbers))

##Explanation: filter() applies the lambda function (lambda x: x % 2 == 0) to each element, keeping only those elements for which the function returns True. In this case, it filters out the odd numbers, leaving only [2, 4].

#Key Differences:>>>>>>>>>>>>>>>>>>>>>>>>>>>

#map(): Applies a function to each element in an iterable and returns a transformed collection.
#Example: Squaring each number.

#reduce(): Applies a function cumulatively to elements in an iterable, reducing them to a single value.
#Example: Summing all numbers.

#filter(): Applies a function that returns a boolean value to filter elements based on a condition.
#Example: Keeping only even numbers.

#Summary:
#Use map() when you want to transform each element of an iterable.
#Use reduce() when you want to combine all elements of an iterable into a single value.
#Use filter() when you want to select elements that satisfy a certain condition.
```

```python
##question 1 > Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

#Here’s a practical Python function that takes a list of numbers and returns the sum of all even numbers in the list using a simple loop. This is an efficient and straightforward approach.

#Python Function:

def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:  # Check if the number is even
            total += num   # Add even number to the total sum
    return total

# Example usage:
numbers = [47, 11, 42, 13, 8, 24]
result = sum_even_numbers(numbers)
print("Sum of even numbers:", result)

#Explanation:
#Looping through the list: The function iterates over each number in the list.
#Check if even: Inside the loop, we check if the number is even by using the condition num % 2 == 0.
#Accumulate sum: If the number is even, it gets added to the total sum.
#Return the sum: After the loop finishes, the function returns the sum of all even numbers.

#Example Output:
#Sum of even numbers: 74

# This solution is simple and efficient for summing the even numbers in a list.



```

```python
## question 2>>>> Create a Python function that accepts a string and returns the reverse of that string

#Here’s a practical Python function that accepts a string and returns the reverse of that string using Python's slicing feature:

#Python Function to Reverse a String:

def reverse_string(s):
    return s[::-1]

# Example usage:
input_string = "hello"
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)

#Explanation:
#Slicing ([::-1]): This is a concise and efficient way to reverse a string in Python. The slice [::-1] means:
#: means the whole string.
#-1 indicates the step, which tells Python to traverse the string from end to start.
#The function returns the reversed version of the input string.

#Example Output:
#Reversed string: olleh

#This approach is simple, efficient, and widely used for reversing strings in Python.
```

```python
## question 3> Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

#Here's a practical Python function that takes a list of integers and returns a new list containing the squares of each number:

#Python Function to Square Each Number:
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example usage:
numbers = [1, 2, 3, 4, 5]
squared_numbers = square_numbers(numbers)
print("Squared numbers:", squared_numbers)


#Explanation:
#List Comprehension: [num ** 2 for num in numbers]
#This iterates over each number in the numbers list and squares it (num ** 2).
#The result is a new list where each element is the square of the corresponding element from the original list.

#Example Output:

#Squared numbers: [1, 4, 9, 16, 25]

#This is a concise and efficient solution to square each element in a list.
```

```python

## question 4 >> Write a Python function that checks if a given number is prime or not from 1 to 200.

#Here’s a practical Python function that checks if a given number is prime, specifically for numbers between 1 and 200:

#Python Function to Check Prime Number:

def is_prime(n):
    if n <= 1:  # Numbers less than or equal to 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):  # Check divisibility from 2 to √n
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num)

#Explanation:
#Check if n is less than or equal to 1: Numbers 1 and below are not prime.
#Efficient checking using int(n**0.5) + 1:
#You only need to check divisibility up to the square root of n because if a number n is divisible by some number p, then n = p * q, and if both p and q are greater than √n, their product would exceed n. Hence, we only check divisibility up to √n.
#Return True if no divisors are found: If the number is not divisible by any number between 2 and √n, it is prime.

#Key Points:
#Efficiency: The function only checks up to the square root of the number, which makes it more efficient than checking all numbers from 2 to n-1.
#Practicality: This approach works well for numbers in the range 1 to 200, and can be easily extended to larger ranges if necessary.
```

```python
## question 5 >> Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

#Here's a practical implementation of an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms. This class implements both the __iter__() and __next__() methods, making it a proper iterator.

#Fibonacci Iterator Class:

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms in the Fibonacci sequence to generate
        self.current = 0  # Current term index
        self.a, self.b = 0, 1  # First two terms of the Fibonacci sequence

    def __iter__(self):
        return self  # The iterator object is itself
    
    def __next__(self):
        if self.current >= self.terms:  # Stop iteration after reaching the specified terms
            raise StopIteration
        
        # Generate the next Fibonacci number
        fibonacci_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update the Fibonacci sequence
        self.current += 1  # Move to the next term
        
        return fibonacci_number

# Example usage:
fibonacci_gen = FibonacciIterator(10)  # Generate first 10 terms of the Fibonacci sequence
for num in fibonacci_gen:
    print(num)

#Explanation:
#1 __init__(self, terms):

#The constructor takes the number of terms to generate as an argument.
#self.a, self.b represent the first two terms of the Fibonacci sequence (0 and 1).
#self.current keeps track of how many terms have been generated so far.

#__iter__(self):
#This method returns the iterator object itself. It is required for the object to be iterable.

#__next__(self):
#This method generates the next Fibonacci number.
#It raises a StopIteration exception when the number of terms (self.terms) is reached.
#The Fibonacci sequence is generated by updating self.a and self.b with the next pair of terms, where each new term is the sum of the previous two.

#Key Points:
#Iterator Class: The FibonacciIterator class is an iterator because it implements __iter__() and __next__().
#Efficiency: It generates Fibonacci numbers on-the-fly (lazy evaluation), which means it does not require storing all terms in memory at once.
#Flexibility: You can specify how many terms of the Fibonacci sequence you want to generate when you create the iterator.
```

```python
## question 6 >> Write a generator function in Python that yields the powers of 2 up to a given exponent.

#Here’s a practical implementation of a Python generator function that yields the powers of 2 up to a given exponent:

#Generator Function for Powers of 2:
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):
    print(power)

#Explanation:

#1..powers_of_2(exponent):
#The function takes an exponent as an argument, which determines the highest power of 2 to be yielded.
#The for loop iterates over the range from 0 to the given exponent (exponent + 1 to include the final power).
#For each i, 2 ** i calculates the power of 2 for the current iteration and yields it.

#2..Generator (yield):
#The yield keyword makes the function a generator. Each time yield is called, the function pauses and returns the value, then resumes execution from where it left off on the next iteration.

#Key Points:
#Lazy Evaluation: The generator yields one power of 2 at a time, so it doesn't need to store all the values in memory.
#Efficiency: It’s a memory-efficient way to generate the powers of 2, especially for large exponents, since it produces each value only when needed.
```

```python
## question 8 >>  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

#Here’s a practical example of how to use a lambda function to sort a list of tuples based on the second element of each tuple:

#Python Code:

# List of tuples
tuples_list = [(1, 3), (2, 1), (4, 2), (3, 4)]

# Sorting the list of tuples based on the second element using a lambda function
sorted_tuples = sorted(tuples_list, key=lambda x: x[1])

# Printing the sorted list
print(sorted_tuples)

#Explanation:
#sorted() function: This built-in function is used to sort an iterable (in this case, a list of tuples).
#key=lambda x: x[1]: The key argument specifies a function to be called on each element before making comparisons. In this case, the lambda function takes each tuple x and returns the second element (x[1]), which is used for sorting.

#Key Points:
#The lambda function (lambda x: x[1]) allows sorting based on the second element of each tuple.
#The sorted() function returns a new list sorted in ascending order based on the provided key function.
```

```python
## question 9 >> Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

#Here's a practical Python program that uses the map() function to convert a list of temperatures from Celsius to Fahrenheit:

#Python Code:

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40, 50]

# Conversion function: Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion function to each element in the list
fahrenheit_temps = map(celsius_to_fahrenheit, celsius_temps)

# Convert the map object to a list and print the result
print(list(fahrenheit_temps))

#Explanation:
#1 celsius_to_fahrenheit(celsius): This function takes a temperature in Celsius and returns the equivalent temperature in Fahrenheit using the formula:
#         Fahrenheit=(Celsius× 9/5)+32

#2 map(celsius_to_fahrenheit, celsius_temps): The map() function applies the celsius_to_fahrenheit function to each element in the celsius_temps list. It returns an iterator that produces the Fahrenheit temperatures.
#3 list(fahrenheit_temps): Since map() returns an iterator, we convert it to a list to see all the converted temperatures.

#Key Points:
#map() is a clean and efficient way to apply a function to each item in an iterable.
#The result is a generator-like object, which can be converted to a list (or any other iterable) for further use.

```

```python
## question 11 >>>>>>>>>

#To solve this problem using lambda and map, we can process the given data to compute the required values (order number and the adjusted total cost). If the total cost is smaller than 100, we will increase it by 10€.

#Problem Breakdown:
#1 We have four sublists:
#Order numbers: [34587, 98762, 77226, 88112]
#Book titles and authors: ['learning python mark lutz', 'programming python mark lutz', 'head first python paul barry', 'einfuhruung in python3 bernd klein']
#Quantities: [4, 5, 3, 3]
#Price per item: [40.95, 56.80, 32.95, 24.99]

#2>.We need to compute the total cost for each order by multiplying the quantity by the price per item.
#3>>If the total cost is less than 100€, add 10€ to it.
#4Return a list of 2-tuples, where each tuple contains:
#Order number
#Total cost (adjusted if needed)

#Solution:
#We'll use the map() function to iterate over the data and calculate the required results.

#Python Program:

# Data: Order numbers, book titles, quantities, and prices
order_numbers = [34587, 98762, 77226, 88112]
book_titles = ['learning python mark lutz', 'programming python mark lutz', 'head first python paul barry', 'einfuhruung in python3 bernd klein']
quantities = [4, 5, 3, 3]
prices = [40.95, 56.80, 32.95, 24.99]

# Lambda function to calculate the total cost and apply the 10€ adjustment if necessary
calculate_total = lambda order, quantity, price: (order, (quantity * price + 10) if quantity * price < 100 else quantity * price)

# Use map() to apply the lambda function to the data
result = list(map(lambda data: calculate_total(data[0], data[1], data[2]), zip(order_numbers, quantities, prices)))

# Print the result
print(result)


#Explanation:
#Data: We have four separate lists: order_numbers, book_titles, quantities, and prices.
#calculate_total: This lambda function calculates the total cost of an order by multiplying the quantity and price. If the total cost is less than 100€, we add 10€ to it.
#zip(): We use zip() to combine the three lists (order_numbers, quantities, and prices) into a single iterable of tuples. Each tuple contains an order number, a quantity, and a price.
#map(): The map() function applies the lambda function to each tuple created by zip(). This processes each order, calculates the total cost (adjusted if needed), and returns a tuple of the order number and the computed total cost.
#Result: The final result is a list of tuples, where each tuple contains an order number and the adjusted total cost.

#Key Points:
#Lambda Functions: We use a lambda function to calculate the total cost and adjust it if necessary.
#map(): The map() function applies the lambda to each element, simplifying the code and making it more concise.
#zip(): The zip() function combines the three lists into a single iterable of tuples, which makes it easy to process corresponding elements from each list.

#This approach is clean and efficient for processing the given data and performing the necessary computations.
```
